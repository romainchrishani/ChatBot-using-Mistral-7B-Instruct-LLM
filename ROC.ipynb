```python
import pandas as pd
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel('/content/ROC_1.xlsx')

# Convert 'Yes'/'No' in 'TrueLabels' to 1/0
y_true = df['TrueLabels'].map({'Yes': 1, 'No': 0}).tolist()
print(y_true)

# Assuming 'Scores' are already in the correct format
y_scores = df['Scores'].tolist()
print(y_scores)

# Handling 'inf' values if they exist in y_scores
y_scores = [score if score != float('inf') else max(y_scores) + 1 for score in y_scores]

# Compute the ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(y_scores, bins=30, edgecolor='black')
plt.title('Histogram of Prediction Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```
